```python
import pandas as pd
import time
import os
from datetime import datetime
```

```python
movies_names = ['Movie_Id', 'Name_Movie']
```

```python
movies = pd.read_csv('/content/movies.csv', sep=';', header=None, names=movies_names)
```

Remover Parenteses

```python
movies['Name_Movie'] = movies['Name_Movie'].str[1:-1]
movies['Name_Movie']
```

**1.2. Qual Ã© o nome dos 5 filmes com melhor mÃ©dia de avaliaÃ§Ã£o?**

Realizando a mÃ©dia e em seguida ordenando de forma decrecente, poderemos verificar os top 5

**1.4. Quantos filmes que possuem avaliaÃ§Ã£o maior ou igual a 4.7, considerando apenas os filmes avaliados na Ãºltima data de avaliaÃ§Ã£o do dataset?**

```python
new_df.Date.max()
```

```python
best_df
```

```python
best_df['Name_Movies'].nunique()
```

```python
nome_filmes_best_df = best_df.Name_Movies.unique()
```

```python
name_best_df = pd.DataFrame(nomes_filmes_best_df)
name_best_df = name_best_df.rename(columns={0: "Name_Movies"})
name_best_df 
```
